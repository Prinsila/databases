```python
# prompt: change the linkedin ids from 1 to up

import pandas as pd

df = pd.read_excel('/content/Crazy frogs.xlsx')

# Check for missing values in each column
print(df.isnull().sum())

# Remove rows with missing values (if any)
df = df.dropna()

# Check data types and convert if needed
print(df.dtypes)

# Assuming 'linkedin_id' is a numeric column, but needs to be converted to int.
if not pd.api.types.is_numeric_dtype(df['linkedin_id']):
    try:
        df['linkedin_id'] = pd.to_numeric(df['linkedin_id'], errors='coerce').astype('Int64') # Use 'Int64' for nullable integers
    except (ValueError, TypeError):
        print("Could not convert 'linkedin_id' to numeric. Check the column for invalid values.")

# Remove duplicates
df = df.drop_duplicates()

# Check for inconsistent data (e.g. capitalization) in categorical columns
for col in ['job_title', 'contract', 'mode', 'city', 'country', 'company', 'sector', 'job_posting_title']:
    if df[col].dtype == 'object' or pd.api.types.is_categorical_dtype(df[col]):
        df[col] = df[col].str.lower()

# Reset the index after dropping rows
df = df.reset_index(drop=True)

# Now, modify the 'linkedin_id' column to sequential numbers
df['linkedin_id'] = range(1, len(df) + 1)


# Display the cleaned DataFrame
df
```

```python
# prompt: drop the column mode, city, and company

import pandas as pd

df = pd.read_excel('/content/Crazy frogs.xlsx')

# Check for missing values in each column
print(df.isnull().sum())

# Remove rows with missing values (if any)
df = df.dropna()

# Check data types and convert if needed
print(df.dtypes)

# Assuming 'linkedin_id' is a numeric column, but needs to be converted to int.
if not pd.api.types.is_numeric_dtype(df['linkedin_id']):
    try:
        df['linkedin_id'] = pd.to_numeric(df['linkedin_id'], errors='coerce').astype('Int64') # Use 'Int64' for nullable integers
    except (ValueError, TypeError):
        print("Could not convert 'linkedin_id' to numeric. Check the column for invalid values.")

# Remove duplicates
df = df.drop_duplicates()

# Check for inconsistent data (e.g. capitalization) in categorical columns
for col in ['job_title', 'contract', 'mode', 'city', 'country', 'company', 'sector', 'job_posting_title']:
    if df[col].dtype == 'object' or pd.api.types.is_categorical_dtype(df[col]):
        df[col] = df[col].str.lower()

# Reset the index after dropping rows
df = df.reset_index(drop=True)

# Now, modify the 'linkedin_id' column to sequential numbers
df['linkedin_id'] = range(1, len(df) + 1)

# Drop the specified columns
df = df.drop(columns=['mode', 'city', 'company'])

# Display the cleaned DataFrame
df
```
